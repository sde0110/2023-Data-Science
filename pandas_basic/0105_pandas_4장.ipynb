```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/Users/shindongeun/Documents/sample_data/시도별 전출입 인구수.xlsx',header =0)
```

```python
df.head()
```

```python
# NaN을 앞 값으로 채움
df = df.fillna(method='ffill')
#NaN을 특정 값으로 채우려면 fillna(0)
#뒤에 있는 값으로 채우려면 fillna(method='bfill')

df.head(4)
```

```python
# 조건 연산자를 이용한 인덱싱 (->서울에서 나가서 다른 지역으로 간 인구))
df_seoul = df[(df['전출지별']=='서울특별시')&
              (df['전입지별']!= '서울특별시')]
df_seoul.head()
```

```python
# 전출지가 다 서울이라서 필요없어서 삭제
del df_seoul['전출지별']
df_seoul.head()
```

```python
df_seoul.rename({'전입지별':'전입지'},axis=1,inplace=True)
# 전입지별의 이름을 전입지로 변경
```

```python
# 전입지를 인덱스로 설정
df_seoul = df_seoul.set_index('전입지')
# = 쓰기 싫으면 맨 뒤에 ,inplace=True 옵션값 넣기
df_seoul.head()
```

```python
# 전입지가 '경기도'인 행 데이터를 선택해서 sr_one에 저장
# 문자열로 찾을거니까 iloc X, loc O
sr_one = df_seoul.loc['경기도']
sr_one #시리즈 원
```

```python
sr_one.values
```

## 그래프 꾸미기
    x축 눈금이 겹칠 때 해결방법
~~~
figure() 함수로 가로 사이즈를 더 크게 설정
xticks()함수를 활용하여 x축 눈금 라벨을 회전
ppt처럼 화살표를 표시할 수도 있다 !
~~~

```python
#애플 폰트 설정
plt.rc('font', family='AppleGothic') 

#스타일 서식 지정
plt.style.use(['fivethirtyeight'])

#그래프 꾸미기
plt.figure(figsize=(14,5)) # 캔버스 사이즈 키우기 
plt.xticks(rotation=60) # 각도 조절 

#plt.plot(x축값, y축값) (연도, 인구)
plt.plot(sr_one.index,sr_one.values,marker='h',markersize=5, color='brown',linewidth=3) #marker 는 점 모양

#차트 제목 추가
plt.title( "'서울 -> 경기' 인구 이동",size = 30)
#축 이름 추가
plt.xlabel('기간',size=20)
plt.ylabel('이동 인구수',size=20)

#범례 표시
plt.legend(labels=['서울->경기'],loc='best',fontsize=15) #best option은 적당한 곳에 놔두라는 뜻


# plt 객체에서 정의한 plt를 보여줘
plt.show()
```

## 화면 분할하여 그래프 여러개 그리기
~~~
axe 활용

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,1,1) #add_subplot(행의 갯수, 열의 갯수, n번째 위치)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(sr_one,'o',markersize=5) #sr_one만 적으면 x축 : 인덱스, y축 : 값으로 자동 지정됨
ax2.plot(sr_one, marker='^', markerfacecolor='green',markersize=5,color='olive',linewidth=2,label='서울->경기')
ax2.legend(loc='best')

#xy축 범위 지정 (최소값, 최대값)
ax1.set_ylim(50000,800000)
ax2.set_ylim(50000,800000)

# 축 눈금 라벨 지정 및 회전
ax1.set_xticklabels(sr_one.index,rotation=75)
ax2.set_xticklabels(sr_one.index,rotation=75)
plt.show()
```

```python
# 서울에서 '충청, 경상, 강원'으로 이동한 인구 데이터만 선택
col_years_list = list(range(1970,2018))
col_years = []
for i in col_years_list:
    col_years.append(str(i))
col_years[:10]
```

```python
df3 = df_seoul.loc[['충청남도','경상북도','강원도'],col_years]
df3.head()
```

```python
#애플 폰트 설정
plt.rc('font', family='AppleGothic') 

#스타일 서식 지정
plt.style.use(['ggplot'])

#그래프 객체 생성 (figure에 1개의 서브 플롯을 생성)
fig = plt.figure(figsize = (20,5))
ax = fig.add_subplot(1,1,1)

#axe 객체에 plot 함수로 그래프 출력
ax.plot(col_years, df3.loc['충청남도',:],marker='o',markerfacecolor='green',
       markersize=10,color='olive',linewidth=2,label='서울->충남')
ax.plot(col_years, df3.loc['경상북도',:],marker='o',markerfacecolor='#7e10cc',
       markersize=10,color='skyblue',linewidth=2,label='서울->경북')
ax.plot(col_years, df3.loc['강원도',:],marker='o',markerfacecolor='red',
       markersize=10,color='magenta',linewidth=2,label='서울->강원')

#범례 표시
plt.legend(loc='best',fontsize=15) #best option은 적당한 곳에 놔두라는 뜻

plt.title( "'서울 -> 충남,경북,강원' 인구 이동",size = 20)

#축 이름 추가
plt.xlabel('기간',size=12)
plt.ylabel('이동 인구수',size=12)

#축 눈금 라벨 지정 및 회전
ax.set_xticklabels(col_years,rotation=90)

#축 눈금 라벨 크기
ax.tick_params(axis="x",labelsize=10)
ax.tick_params(axis="y",labelsize=10)


# plt 객체에서 정의한 plt를 보여줘
plt.show()
```
