```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/Users/shindongeun/Documents/sample_data/시도별 전출입 인구수.xlsx',header =0)
```

```python
df.head()
```

```python
# NaN을 앞 값으로 채움
df = df.fillna(method='ffill')
#NaN을 특정 값으로 채우려면 fillna(0)
#뒤에 있는 값으로 채우려면 fillna(method='bfill')

df.head(4)
```

```python
# 조건 연산자를 이용한 인덱싱 (->서울에서 나가서 다른 지역으로 간 인구))
df_seoul = df[(df['전출지별']=='서울특별시')&
              (df['전입지별']!= '서울특별시')]
df_seoul.head()
```

```python
# 전출지가 다 서울이라서 필요없어서 삭제
del df_seoul['전출지별']
df_seoul.head()
```

```python
df_seoul.rename({'전입지별':'전입지'},axis=1,inplace=True)
# 전입지별의 이름을 전입지로 변경
```

```python
# 전입지를 인덱스로 설정
df_seoul = df_seoul.set_index('전입지')
# = 쓰기 싫으면 맨 뒤에 ,inplace=True 옵션값 넣기
df_seoul.head()
```

```python
# 전입지가 '경기도'인 행 데이터를 선택해서 sr_one에 저장
# 문자열로 찾을거니까 iloc X, loc O
sr_one = df_seoul.loc['경기도']
sr_one #시리즈 원
```

```python
sr_one.values
```

## 그래프 꾸미기

~~~
figure() 함수로 가로 사이즈를 더 크게 설정

```python
#애플 폰트 설정
plt.rc('font', family='AppleGothic') 

#스타일 서식 지정
plt.style.use(['fivethirtyeight'])

#그래프 꾸미기
plt.figure(figsize=(14,5)) # 캔버스 사이즈 키우기 
plt.xticks(rotation=60) # 각도 조절 

#plt.plot(x축값, y축값) (연도, 인구)
plt.plot(sr_one.index,sr_one.values,marker='h',markersize=5, color='brown',linewidth=3) #marker 는 점 모양

#차트 제목 추가
plt.title( "'서울 -> 경기' 인구 이동",size = 30)
#축 이름 추가
plt.xlabel('기간',size=20)
plt.ylabel('이동 인구수',size=20)

#범례 표시
plt.legend(labels=['서울->경기'],loc='best',fontsize=15) #best option은 적당한 곳에 놔두라는 뜻


# plt 객체에서 정의한 plt를 보여줘
plt.show()
```

```python
# 서울에서 '충청, 경상, 강원'으로 이동한 인구 데이터만 선택
col_years_list = list(range(1970,2018))
col_years = []
for i in col_years_list:
    col_years.append(str(i))
col_years[:10]
```

```python
import seaborn as sns
#import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
print(titanic.head())
print('*'*60)
print(titanic.info())
```

```python
sns.set_style('darkgrid')
import seaborn as sns
```

```python
#회귀선이 있는 산점도 그래프 -> 모든 점을 표현할 수 있는 하나의 직선 : 회귀선

#그래프 객체 생성( figure에 2개의 서브플롯을 생성)
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

#그래프 그리기 - 선형회귀선 표시 : fit_reg
sns.regplot(x='age',y='fare',data=titanic,
            line_kws={'color':'yellow'},scatter_kws={'color':'coral'},
            ax=ax1,fit_reg = True) 
# 순서대로 <x축, y축, 데이터, axe객체, 선형회귀선>
sns.regplot(x='age',y='fare',data=titanic,
            scatter_kws={'color':'skyblue'},
            ax=ax2,fit_reg = False) 
plt.show()
```

```python
geo_path = '/Users/shindongeun/Documents/sample_data/경기도행정구역경계.json'
try:
    geo_data = json.load(open(geo_path, encoding='utf-8'))
    
except:
    geo_data = json.load(open(geo_path, encoding='utf-8-sig'))
    
# 경기도 지도 만들기
g_map = folium.Map(location=[37.5502,126.982], tiles='Stamen Terrain',
                  zoom_start=9)

# 출력 연도 선택 (2007~2017)
year = '2017'

# chropleth class로 단계구분 표시하기
folium.Choropleth(geo_data=geo_data,
                 data= df[year], columns = [df.index, df[year]],
                 fill_color='BuPu',fill_opacity=0.7, line_opacity=0.3,
                 threshold_scale=[10000,100000,300000,500000,700000],
                 key_on = 'feature.properties.name',
                 ).add_to(g_map)

# 지도를 HTML 파일로 저장하기
g_map.save('/Users/shindongeun/Documents/sample_data/gyonggi_population'
          +year+'.html')
g_map
```
