## 그래프

```python
df = pd.read_excel('/Users/shindongeun/Documents/sample_data/남북한발전전력량.xlsx')
df_ns = df.iloc[[0,5],3:]
df_ns.index = ['South','North']
df_ns.columns = df_ns.columns.map(int)
display(df_ns.head())
```

```python
df_ns.plot() #x축은 index, y축은 column
```

```python
# 선그래프
tdf_ns = df_ns.T
display(tdf_ns.head())
print('-'*40)
tdf_ns.plot()
```

```python
# 막대그래프
tdf_ns.plot(kind='bar')
```

```python
# 히스토그램
df_ns = df.iloc[[0,5],3:]
df_ns.index = ['South','North']
df_ns.columns = df_ns.columns.map(int)
tdf_ns = df_ns.T
tdf_ns.plot(kind='hist')
```

```python
# 산점도그래프
tdf_ns.plot(x = 'weight',y = 'mpg',kind = 'scatter')
#양의 상관관계, 음의 상관관계
# 상관계수가 0에 가까우면 -> 퍼져있다. 낮은 상관도
# -1에 가까우면 -> 좁은형태, 높은 상관도
# 1에 가까우면 -> 양의 형태, 높은 상관도
```
